#### Functions

```python
def read_data_churn(train) -> DataFrame:
    
    """This function automatically reads a dataframe processed 
    with all features in S3 and return this dataframe with
    cid as index
    
    Parameters
    ----------

    Returns
    -------
    Pandas dataframe containing all features
    """

    s3 = boto3.resource('s3')
    if train:
        prefix = config['paths']['dir_main_data'] + 'train_data_1'
    else:
        prefix = config['paths']['dir_main_data'] + 'po_data_1'
    bucket_name = config['paths']['bucket_name']
    my_bucket = s3.Bucket(bucket_name)
    files = my_bucket.objects.filter(Prefix=prefix)

    df_features_churn = pd.DataFrame()
    for file in files:
        if 'SUCCESS' not in file.key:   
            df = pd.read_orc("s3://" + bucket_name + '/' + file.key)
            df_features_churn = df_features_churn.append(df)
    df_features_churn.index = df_features_churn['cid']
    df_features_churn.index.name = 'cid'
    
    return df_features_churn

def translate_value(df_features_churn: DataFrame) -> DataFrame:
    
    """This function translate the column value to numeric
    
    Parameters
    ----------
    df_features_churn: Pandas dataframe with features
    
    Returns
    -------
    Pandas dataframe with new columns containing numerics values
    translated from value column
    """
    
    df_features_churn['value_cat'] = df_features_churn['value'].replace({'LOW': 1, 'MID': 2, 'HIGH': 3})

    return df_features_churn

def extract_trend(df: DataFrame) -> DataFrame:
    
    """This function creates two new columns from a global lisit
    of yearly variables. It computes the trends between year 1 and 2
    and between year 2 and 3
    
    Parameters
    ----------
    df: Pandas dataframe with features
    
    Returns
    -------
    Pandas dataframe with new columns containing trends between columns
    """
    
    for col in config['variables']['columns_trend']:
        df['trend1_' + col] = ((df[col + '_year1']-df[col + '_year2'])/df[col + '_year2']).replace(np.inf, 1).replace(-np.inf, 1).replace(np.nan, 0)
        df['trend2_' + col] = ((df[col + '_year2']-df[col + '_year3'])/df[col + '_year3']).replace(np.inf, 1).replace(-np.inf, 1).replace(np.nan, 0)
    
    # df['trend_q1q2'] = ((df['gross_revenue_and_ancillaries_year1_q1']-df['gross_revenue_and_ancillaries_year1_q2'])/df['gross_revenue_and_ancillaries_year1_q2']).replace(np.inf, 1).replace(-np.inf, 1)
    # df['trend_q2q3'] = ((df['gross_revenue_and_ancillaries_year1_q2']-df['gross_revenue_and_ancillaries_year1_q3'])/df['gross_revenue_and_ancillaries_year1_q3']).replace(np.inf, 1).replace(-np.inf, 1)
    # df['trend_q3q4'] = ((df['gross_revenue_and_ancillaries_year1_q3']-df['gross_revenue_and_ancillaries_year1_q4'])/df['gross_revenue_and_ancillaries_year1_q4']).replace(np.inf, 1).replace(-np.inf, 1)
    
    return df

def auxiliar_type_conversion(df_features_churn: DataFrame) -> DataFrame:
    
    """This function converts types from feature dataframe
    
    Parameters
    ----------
    df_features_churn: Pandas dataframe with features
    
    Returns
    -------
    Pandas dataframe with casted columns
    """
    
    df_features_churn['ff_tier'] = df_features_churn['ff_tier'].astype(float)
    df_features_churn['purchases_2015'] = df_features_churn['purchases_2015'].astype(float)
    df_features_churn['purchases_2016'] = df_features_churn['purchases_2016'].astype(float)
    df_features_churn['purchases_2017'] = df_features_churn['purchases_2017'].astype(float)
    
    return df_features_churn

def feature_encoder_train(X_train: DataFrame, X_test: DataFrame, X_val: DataFrame) -> DataFrame:
    
    
    # Binning
    Kbin_age = KBinsDiscretizer(n_bins=10,encode='ordinal',strategy='quantile')
    trf_cat = ColumnTransformer([('age_bin', Kbin_age, ['age'])])
    X_train.loc[X_train['age'].notnull(), 'age_bin'] = trf_cat.fit_transform(X_train[X_train['age'].notnull()])
    X_test.loc[X_test['age'].notnull(), 'age_bin'] = trf_cat.transform(X_test[X_test['age'].notnull()])
    X_val.loc[X_val['age'].notnull(), 'age_bin'] = trf_cat.transform(X_val[X_val['age'].notnull()])
    # Target
    tg_encoder = TargetEncoder()
    X_train['most_od_flown_tg'] = tg_encoder.fit_transform(X_train['most_od_flown'],
                                                    X_train[config['variables']['target']])
    X_test['most_od_flown_tg'] = tg_encoder.transform(X_test['most_od_flown'],
                                                   X_test[config['variables']['target']])
    X_val['most_od_flown_tg'] = tg_encoder.transform(X_val['most_od_flown'],
                                                  X_val[config['variables']['target']])
    # Save binning encoder
    trf_cat_dump = pickle.dumps(trf_cat)
    s3_resource.Object('iberia-data-lake', f"customer/churn_model_v2/models/cat_encoder.pkl").put(
        Body=trf_cat_dump
    )
    # Save binning encoder
    tg_encoder_dump = pickle.dumps(tg_encoder)
    s3_resource.Object('iberia-data-lake', f"customer/churn_model_v2/models/target_encoder.pkl").put(
        Body=tg_encoder_dump
    )
    
    return X_train, X_test, X_val

def feature_encoder_predict(X: DataFrame) -> DataFrame:
    
    # Read binning encoder
    trf_cat = (
        s3_resource.Bucket('iberia-data-lake')
        .Object("customer/churn_model_v2/models/cat_encoder.pkl")
        .get()
    )
    trf_cat = pickle.loads(trf_cat["Body"].read())
    
    # Read target encoder
    tg_encoder = (
        s3_resource.Bucket('iberia-data-lake')
        .Object("customer/churn_model_v2/models/target_encoder.pkl")
        .get()
    )
    tg_encoder = pickle.loads(tg_encoder["Body"].read())

    # Impute new data
    X.loc[X['age'].notnull(), 'age_bin'] = trf_cat.transform(X[X['age'].notnull()])
    X['most_od_flown_tg'] = tg_encoder.transform(X['most_od_flown'],
                                                 X[config['variables']['target']])

    return X
    
def plot_roc_curve(fpr: list, tpr: list) -> None:
    
    """This function plots the ROC Curve
    
    Parameters
    ----------
    fpr: array with false positive rates (fpr) with different thresholds
    tpr: array with true positive rates (tpr) with different thresholds
    
    Returns
    -------
    None
    """

    plt.plot(fpr, tpr, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()

def compute_auc_roc(final_model_churn: object, X_test_churn: DataFrame, y_test_churn: object, y_pred_churn: object) -> None:

    """This function compute AUC and plot ROC Curve
    
    Parameters
    ----------
    final_model_churn: model already trained
    X_test_churn: pandas dataframe with test features
    y_test_churn: pandas series with real labels for test
    y_pred_churn: pandas series with predictions for test
    
    Returns
    -------
    None
    """

    y_test_prob_churn = final_model_churn.predict_proba(X_test_churn)
    fpr, tpr, thresholds = metrics.roc_curve(y_test_churn, y_test_prob_churn[:,1])
    print(metrics.roc_auc_score(y_test_churn, y_pred_churn))
    optimal_idx = np.argmax(tpr - fpr)
    optimal_threshold = thresholds[optimal_idx]
    print("Threshold value is:", optimal_threshold)
    plot_roc_curve(fpr, tpr)
    
def confusion_matrix_comp(cm: object) -> None:
    
    """This function plots a confussion matrix
    
    Parameters
    ----------
    cm: array containing confussion matrix values
    
    Returns
    -------
    None
    """
    
    df_cm = pd.DataFrame(cm, range(2), range(2))
    sns.set(font_scale=1) # for label size
    sns.heatmap(df_cm, annot=True, annot_kws={"size": 10}, fmt='g') # font size
    plt.title('Confusion Matrix')
    plt.show()
```

#### Reading data

```python
def read_data_and_fe_process(train: bool=False) -> object:
    
    """This reads churn daa from s3 (last partition), creates some features
    and preprocess some columns in order to finally divide this dataframe
    between churn and cltv (with no churners)
    
    Parameters
    ----------
    None
    
    Returns
    -------
    Two dataframes containing all data preprocess (churn) and cltv data
    preprocess (only not churners)
    """

    # Read data
    df = read_data_churn(train)

    # Extract trend
    df = extract_trend(df)

    # Last purchase new features
    df['last_purchase1_2'] = df['last2_purchase'] - df['last_purchase']
    df['last_purchase2_3'] = df['last3_purchase'] - df['last2_purchase']
    
    # Categorical type
    df['most_od_flown'] = df['most_od_flown'].astype('category')

    # Cast variables
    df = auxiliar_type_conversion(df)

    return df

def mice_null_imputation_train(X_train: DataFrame, X_test: DataFrame, X_val: DataFrame) -> DataFrame:
    
    # Define MICE kernel
    kernel = mf.ImputationKernel(
        data=X_train.drop(['most_od_flown', 'churn'], axis=1),
        variable_schema=config['variables']['input_missing_mice'],
        # data_subset=0.5,
        datasets=1
    )
    # Train imputer KNN
    kernel.mice(3,verbose=True)
    # Impute new data
    X_train_im = kernel.impute_new_data(X_train.drop(['most_od_flown', 'churn'], axis=1)).complete_data(0)
    X_train_im[['most_od_flown', 'churn']] = X_train[['most_od_flown', 'churn']]
    X_test_im = kernel.impute_new_data(X_test.drop(['most_od_flown', 'churn'], axis=1)).complete_data(0)
    X_test_im[['most_od_flown', 'churn']] = X_test[['most_od_flown', 'churn']]
    X_val_im = kernel.impute_new_data(X_val.drop(['most_od_flown', 'churn'], axis=1)).complete_data(0)
    X_val_im[['most_od_flown', 'churn']] = X_val[['most_od_flown', 'churn']]

    # Save KNN imputer
    kernel_dump = pickle.dumps(kernel)
    s3_resource.Object('iberia-data-lake', f"customer/churn_model_v2/models/mice_imputer.pkl").put(
        Body=kernel_dump
    )

    return X_train_im, X_test_im, X_val_im
    
def mice_null_imputation_predict(X: DataFrame) -> DataFrame:
    
    # Read Kernel
    kernel = (
        s3_resource.Bucket('iberia-data-lake')
        .Object("customer/churn_model_v2/models/mice_imputer.pkl")
        .get()
    )
    kernel = pickle.loads(kernel["Body"].read())

    # Impute new data
    X_im = kernel.impute_new_data(X.drop(['most_od_flown', 'churn'], axis=1)).complete_data(0)
    X_im[['most_od_flown', 'churn']] = X[['most_od_flown', 'churn']]

    return X_im
        

def split_train_val_test(X: DataFrame) -> object:
    
    """This function uses the pandas dataframe features, divide between
    target and features and splits this data between:
    - X_train_churn, y_train_churn: training for churn
    - X_test_churn, y_test_churn: testing for churn
    - X_final_test, y_final_test: final test for churn and cltv (unseen data)
    
    Parameters
    ----------
    df: pandas dataframe with churn features and label
    
    Returns
    -------
    All dataframes for training and testing, a set of data randomly sampled
    from churn to introduce to cltv model (in order to have some zero 
    predictions) and those ids selected for this purpose
    """
    
    features_target_fin = (config['variables']['cat_features'] + \
                           config['variables']['binary_features'] + \
                           config['variables']['discrete_features'] + \
                           config['variables']['cont_features'] + \
                           [config['variables']['target']])
    X = X[features_target_fin]
    # Split between train, validation and test
    X_train, X_test = train_test_split(X, test_size = config['train']['test_size'],
                                       random_state=1)
    X_test, X_val = train_test_split(X_test, test_size = config['train']['val_test_size'],
                                     random_state=1)

    return X_train, X_test, X_val

def shap_values_extraction_churn(final_model_churn: object, X_test_churn: DataFrame) -> object:
    
    """This function uses churn model and test data to predict
    shap values in order to obtain interpretability from the model

    Parameters
    ----------
    final_model_churn: model churn
    X_test_churn: pandas dataframe with test features
    
    Returns
    -------
    Shap values for churn
    """
    explainer = shap.TreeExplainer(final_model_churn)
    shap_values = explainer.shap_values(X_test_churn)
    
    shap.summary_plot(shap_values, X_test_churn, feature_names=X_test_churn.columns, max_display=20)
    
    return shap_values

```

## Training model

```python
with open('config.yml') as file:
    config = yaml.safe_load(file)
s3_resource = boto3.resource("s3")
# Read data
df_features_churn = read_data_and_fe_process(train=True)
# Divide train and test
X_train, X_test, X_val = split_train_val_test(df_features_churn)
# Impute missing values
# X_train, X_test, X_val = mice_null_imputation_train(X_train, X_test, X_val)
# Features encoder
X_train, X_test, X_val = feature_encoder_train(X_train, X_test, X_val)
# list features
features_target_fin = (config['variables']['cat_features'] + \
                       config['variables']['binary_features'] + \
                       config['variables']['discrete_features'] + \
                       config['variables']['cont_features'] + \
                       config['variables']['extra_features'])
# Eliminate features
for feat in config['variables']['eliminate_features']:
    features_target_fin.remove(feat)
estimator, parameters = tutils.get_model_and_params(
            model_name=config['train']['model_name']
        )
model_optimizer = tutils.HPTClassifier(
    clf=estimator,
    params=parameters,
    train_set=X_train,
    validation_set=X_val,
    feature_cols=features_target_fin,
    target_col=config['variables']['target'],
    opt_metric=config['train']['opt_metric']
)

hpt_study = create_study(direction="maximize")
hpt_study.optimize(model_optimizer, n_trials=config['train']['n_trials'])

optimized_model = estimator(**hpt_study.best_params)
optimized_model.fit(X_train[features_target_fin], X_train['churn'])

optimized_model_dump = pickle.dumps(optimized_model)
s3_resource.Object('iberia-data-lake', f"customer/churn_model_v2/models/model_catboost_v1.pkl").put(
    Body=optimized_model_dump
)

acc = accuracy_score(
    y_true=X_test[config['variables']['target']],
    y_pred=optimized_model.predict(X_test[features_target_fin]),
)
recall = recall_score(
    y_true=X_test[config['variables']['target']],
    y_pred=optimized_model.predict(X_test[features_target_fin]),
)
f1_s = f1_score(
    y_true=X_test[config['variables']['target']],
    y_pred=optimized_model.predict(X_test[features_target_fin]),
)
prec = precision_score(
    y_true=X_test[config['variables']['target']],
    y_pred=optimized_model.predict(X_test[features_target_fin]),
)
```

```python
results = {}
models = ['CatBoostClassifier']
with open('config.yml') as file:
    config = yaml.safe_load(file)
s3_resource = boto3.resource("s3")
# Read data
df_features_churn = read_data_and_fe_process(train=True)
# Divide train and test
X_train, X_test, X_val = split_train_val_test(df_features_churn)
# Impute missing values
X_train, X_test, X_val = mice_null_imputation(X_train, X_test, X_val)
# Features encoder
X_train, X_test, X_val = feature_encoder(X_train, X_test, X_val)
# list features
features_target_fin = (config['variables']['cat_features'] + \
                       config['variables']['binary_features'] + \
                       config['variables']['discrete_features'] + \
                       config['variables']['cont_features'] + \
                       config['variables']['extra_features'])
features_target_fin.remove('age')
for model in models:
    print(f'-----------------------------------########## {model} ##########-----------------------------------')
    if model == 'CatBoostClassifier':
        trials = 100
    else:
        trials = 100
    estimator, parameters = tutils.get_model_and_params(
                model_name=model
            )
    model_optimizer = tutils.HPTClassifier(
                clf=estimator,
                params=parameters,
                train_set=X_train,
                validation_set=X_val,
                feature_cols=features_target_fin,
                target_col='churn',
                opt_metric="accuracy"
            )
    
    hpt_study = create_study(direction="maximize")
    hpt_study.optimize(model_optimizer, n_trials=trials)

    optimized_model = estimator(**hpt_study.best_params)
    optimized_model.fit(X_train[config['variables']['features_churn']], X_train['churn'])
    
    acc = accuracy_score(
                y_true=X_test['churn'],
                y_pred=optimized_model.predict(X_test[config['variables']['features_churn']]),
            )
    recall = recall_score(
                y_true=X_test['churn'],
                y_pred=optimized_model.predict(X_test[config['variables']['features_churn']]),
            )
    f1_s = f1_score(
                y_true=X_test['churn'],
                y_pred=optimized_model.predict(X_test[config['variables']['features_churn']]),
            )
    prec = precision_score(
                y_true=X_test['churn'],
                y_pred=optimized_model.predict(X_test[config['variables']['features_churn']]),
            )
    
    results[model] = {'acc': acc, 'recall': recall, 'precision': precision, 'f1_score': f1_s, 'best_params': hpt_study.best_params}
```
